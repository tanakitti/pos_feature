```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from nltk.corpus import words
import nltk
```

```python
df = pd.read_csv('./output/thai_types_and_keywords.csv', index_col='Unnamed: 0')
df = df.fillna('')
```

```python
kws = []
for words in tqdm(df['keywords']):
    for word in words.split(','):
        kws.append(word)
        
```

```python
def save_keyword():
    # remove duplicate
    words = set(kws)
    words = list(words)
    
    # remove empty
    words.remove('')
    
    
    df = pd.DataFrame(words)
    df.columns = ['word']
    
    # sort by word
    df = df.sort_values(by=['word'])
    
    #save
    df.to_csv('./output/noun.csv')
```

## find_from_csv 
- should have 2 columns ( word and label )
- 1 means good word and 0 means bad word

```python
def find_from_csv(path):
    # load file
    df = pd.read_csv(path)
    
    # fill empty
    df = df.fillna('0')
    
    fal_pos_dic = []
    for index, row in df.iterrows():
        if(row['label'] == '0'):
            fal_pos_dic.append(row['word'])

    return fal_pos_dic
```

```python
path = './data/thai_words_in_zon.csv'
thai_false_pos = find_from_csv(path)
```

## find_noun
- find noun from a english word list ( it should have a column which is "word" )

```python
def find_noun(path):
    
    # read file
    df = pd.read_csv(path)    
    test = df['word']
    
    # find filter unwanttend word
    final = []
    for x in test:
        if x in words.words():
            final.append(x)

    # filter noun
    nouns = []
    for word in final:
        word = [word]
        result = nltk.tag.pos_tag(word)
        if result[0][1] in ['NN','NNS','NNP','NNPS'] and len(result[0][0])>2 :
            nouns.append(result[0][0])

    # get unwanted word
    unwant_words = []
    for word in df['word']:
        if word not in nouns:
            unwant_words.append(word)
    return unwant_words
        
```
