```python
import pandas as pd
```

```python
df = pd.read_csv('./data/stopwords.csv',encoding='utf-8-sig')
df.fillna(0,inplace = True)
df
```

```python
def find_stopwords(df):
    stopwords = []
    
    for index,row in df.iterrows():
        if row['label'] == 0:
            stopwords.append(row['Stopword'])
    return stopwords
    
```

```python
print(find_stopwords(df))
```
