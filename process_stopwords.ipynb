```python
import pandas as pd
```

```python
df = pd.read_csv('./data/stopwords.2.csv',encoding='utf-8-sig')
df.fillna(0,inplace = True)
df
```

```python
def find_stopwords(df):
    stopwords = []
    
    for index,row in df.iterrows():
        if row['label'] == 0:
            stopwords.append(row['Stopword'])
    return stopwords
    
```

```python
stopwords = find_stopwords(df)
```

```python
df_stopwords = pd.DataFrame(stopwords)
```

```python
df_stopwords.columns = ['stopword']
```

```python
df_stopwords.to_csv('./output/stopwords.453.csv')
```
